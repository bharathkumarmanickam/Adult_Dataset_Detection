# preprocessing Technique

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
col = ['workclass','education','martial-status','occupation','relationship','race','sex','native_country']
```

```python
for i in col:
    X_res[i] = le.fit_transform(X_res[[i]])
```

```python
X_res
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
for i in X_res.columns:
    X_res[i] = scaler.fit_transform(X_res[[i]])
```

```python
X_res
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test = train_test_split(X_res,y_res,test_size=0.33,random_state=42)
```

# Logistic Regression 

```python
from sklearn.metrics import accuracy_score
from sklearn.metrics import confusion_matrix
```

```python
from sklearn.linear_model import LogisticRegressionCV
lrc = LogisticRegressionCV()
```

```python
clf = lrc.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

# Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier
tree = DecisionTreeClassifier(max_depth=3)
clf = tree.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
rf = RandomForestClassifier(n_estimators=300)
clf = rf.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

# SVM

```python
from sklearn.svm import SVC
svm = SVC(kernel="rbf")
clf = svm.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```

```python
accuracy_score(y_pred,y_test)
```

```python
confusion_matrix(y_pred,y_test)
```

# KNN

```python
from sklearn.neighbors import KNeighborsClassifier
neigh = KNeighborsClassifier(n_neighbors=3)
clf = neigh.fit(X_train,y_train)
y_pred = clf.predict(X_test)
```
